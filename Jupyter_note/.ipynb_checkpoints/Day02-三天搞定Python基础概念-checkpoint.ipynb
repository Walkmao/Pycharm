## Scipy

### Scipy的结构

- scipy.integrate - >积分和普通微分方程

- scipy.ndimage - >图像处理
- scipy.optimize - >优化和根查找(root finding)
- scipy.special - >特殊功能
- scipy.stats - >统计功能

### 线性代数

```python
import numpy as np
from scipy import linalg

A = np.random.rand(2,2)
b = np.random.randn(2)
x = linalg.solve(A,b)  # A x = b

eigen = linalg.eig(A)  #求特征向量
det = linalg.det(A)  #求行列式

print(A,'\n',eigen,'\n',det)
```

### 数据拟合

```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def func(x, a, b, c):    
    return a * np.exp(-b * x) + c

x = np.linspace(0, 4, 50)

y = func(x, 2.5, 1.3, 0.5)
ydata = y+0.2*np.random.normal(size=len(x))
popt, pcov = curve_fit(func, x, ydata)
plt.plot(x, ydata, 'b*')
plt.plot(x, func(x, popt[0], popt[1], popt[2]),'r-')
plt.title('$f(x)=ae^{-bx}+c$ curve fitting')
```

### 根搜索

```python
import numpy as np
from scipy import optimize

def fun(x):
    return np.exp(np.exp(x)) - x**2

# 通过初始化点0，找到兴趣0 (find zero of fun with initial point 0)
# 通过Newton-Raphson方法 (by Newton-Raphson)
value1 = optimize.newton(fun, 0)

# 通过二分法找到介于(-5,5)之间的 (find zero between (-5,5) by bisection)
value2 = optimize.bisect(fun, -5, 5)
print(value2)
```

## Matplotlib

### 简单制图

```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,201)
y = x**0.5
plt.plot(x,y,'r-')   #默认图

#图形修改
plt.style.use('dark_background')
plt.figure(figsize = (3,3))
plt.plot(x,x**0.3,'r--')  #间断的点
plt.plot(x,x-1,'k-')  #连续的线
plt.plot(x,np.zeros_like(x),'k-')
```

### 多个制图图例标签和标题

```python
x = np.linspace(0,10,201)
plt.figure(figsize = (5,5))

for n in range(2,5):
    y = x**(1/n)
    plt.plot(x,y,label = 'x^(1/'+ str(n) + ')')

#内置样式
plt.legend(loc = 'best')
plt.style.use('dark_background')

#标签
plt.xlabel('X')
plt.ylabel('Y')

#设置坐标轴的显示范围
plt.xlim(-2,10)  
#标题
plt.title('Multi-plot',fontsize = 16)

#刻度
# plt.xticks(color = 'w')
# plt.yticks(color = 'w')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

x,y = np.mgrid[-5:5:100j,-5:5:100j]
z = x**2 + y**2

fig = plt.figure(figsize = (8,6))
ax = plt.axes(projection = '3d')
surf = ax.plot_surface(x,y,z,rstride=1,\
                      cmap=cm.coolwarm,cstride=1,\
                      linewidth=0)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('3D plot of $z = x^2 + y^2$')
```
